## Q3 A car salesman sells on an average 3 cars per week.

# 3.A. What is Probability that in a given week he will sell some cars? [1 Mark]
# 3.B. What is Probability that in a given week he will sell 2 or more but less than 5 cars? [1 Mark]
# 3.C. Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week. [1 Mark]

```python
l=3
```

```python
prob_of_selling_some_cars = 1 - np.exp(-l)
prob_of_selling_two_or_more_but_less_than_five_cars = 0
for p in range(2, 5):
    prob_of_selling_p_cars = np.exp(-l) * (l**p) / np.math.factorial(p)
    prob_of_selling_two_or_more_but_less_than_five_cars=prob_of_selling_two_or_more_but_less_than_five_cars+prob_of_selling_p_cars
p_values = list(range(0, 10))
prob_of_selling_p_cars = []
for p in p_values:
    prob_of_selling_p_cars.append(np.exp(-l) * (l**p) / np.math.factorial(p))
cmprob = np.cumsum(prob_of_selling_p_cars)
plt.plot(p_values, cmprob)
plt.title("3.C - Poissons Distribution")
plt.xlabel("Number of Cars Sold Per Week")
plt.ylabel("Cumulative Probability")


print("3.A - ", prob_of_selling_some_cars)
print("3.B - ", prob_of_selling_two_or_more_but_less_than_five_cars)

plt.show()
```

```python
def find_outliers_IQR(df, column):

  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1

  upper_fence = Q3 + 1.5 * IQR
  lower_fence = Q1 - 1.5 * IQR


  num_outliers = (df[column] > upper_fence).sum()

  return num_outliers

```

```python
operating_companies = PC_Df[PC_Df['OperatingState'] == 'Operating']
closed_companies = PC_Df[PC_Df['OperatingState'] == 'Closed']
operating_companies_funding_mean = operating_companies['Funds_in_million'].mean()
closed_companies_funding_mean = closed_companies['Funds_in_million'].mean()

plt.figure()
plt.bar(['Operating', 'Closed'], [operating_companies_funding_mean, closed_companies_funding_mean])

plt.title('Mean Funding by Operating State')
plt.xlabel('Operating State')
plt.ylabel('Mean Funding (USD)')

plt.show()

```

Clear Significant difference is veisible between Funding raised by operating and closed startups


Writing the null and alternative hypothesis


Null hypothesis: There is no statistically significant difference between the mean Funding_in_million of companies that are still operating and companies that closed down.

Alternative hypothesis: The mean Funding_in_million of companies that are still operating is statistically significantly different from the mean Funding_in_million of companies that closed down.

```python
operating_companies = PC_Df[PC_Df['OperatingState'] == 'Operating']
closed_companies = PC_Df[PC_Df['OperatingState'] == 'Closed']

operating_companies_funding_mean = operating_companies['Funds_in_million'].mean()
operating_companies_funding_std = operating_companies['Funds_in_million'].std()

closed_companies_funding_mean = closed_companies['Funds_in_million'].mean()
closed_companies_funding_std = closed_companies['Funds_in_million'].std()

t_statistic, p_value = ttest_ind(operating_companies['Funds_in_million'], closed_companies['Funds_in_million'], equal_var=False)

if p_value < 0.05:
  print('There is a statistically significant difference between the Funds raised by companies that are still operating and companies that closed down.')
else:
  print('There is no statistically significant difference between the Funds raised by companies that are still operating and companies that closed down.')
```
